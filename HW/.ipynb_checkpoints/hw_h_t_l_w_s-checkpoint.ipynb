## change data for suitable visualisation creating 

```python
df = pd.read_excel('Домашнее_задание_1.xlsx') # load data from excel file
```

```python
df.head(10) # check dataset for inner information
```

```python
df_pivot = df.pivot(index='company', 
                    columns='year', 
                    values='revenue, $K') # reshape dataset for more suitable visualisation
```

```python
df_pivot
```

```python
years_lst = list(df_pivot.columns) # create yeats list for using in cycle
```

```python
# add columns with revenue growth index

for num, elem in enumerate(years_lst):
    if elem != max(years_lst):
        df_pivot[f"GR % {years_lst[num+1]}/{years_lst[num]}"] = df_pivot[years_lst[num+1]]/df_pivot[years_lst[num]] - 1
    else:
        break                                                                     
```

```python
df_pivot
```

```python
# create two dfs for revenue and revenue growth by years
revenue = df_pivot.iloc[:, :8]
revenue_gr = df_pivot.iloc[:, 8:15]
```

```python
# create graph with revenue growth by years by companies

fig, ax = plt.subplots(figsize=(12, 6))

x_ticks = list(range(len(revenue_gr.columns)))

for num, elem in enumerate(df_pivot.index):
    if elem == 'Shmuber':
        ax.plot(df_pivot.iloc[:, 8:].columns, df_pivot.iloc[:, 8:].loc[elem, :], label=elem, linewidth=4)
    else:
        ax.plot(df_pivot.iloc[:, 8:].columns, df_pivot.iloc[:, 8:].loc[elem, :], label=elem)
        
ax.set_title('Anual revenue growth dynamics from 2009 to 2015', fontsize=20)
ax.set_ylabel('Revenue GR index, %', fontsize=16)
    
vals = ax.get_yticks()

ax.set_xticks(x_ticks)
ax.set_yticklabels([f"{x:.0%}" for x in vals])

ax.legend(bbox_to_anchor=(1.27, 0.6), fontsize=16)

plt.show()
```

```python
# create graph with anual revenue growth by companies from 2010

fig, ax = plt.subplots(figsize=(12, 6))

x_lst = list(range(len(df_pivot.iloc[:, 9:].columns)))

for num, elem in enumerate(df_pivot.index):
    x = df_pivot.iloc[:, 9:].columns
    y = df_pivot.iloc[:, 9:].loc[elem, :]
    if elem == 'Shmuber':
        ax.plot(x, y, label=elem, linewidth=4)
        z = np.polyfit(x_lst, y, 1)
        p = np.poly1d(z)
        ax.plot(x, p(x_lst), linestyle='--')
    else:
        ax.plot(df_pivot.iloc[:, 9:].columns, df_pivot.iloc[:, 9:].loc[elem, :], label=elem)
    
vals = ax.get_yticks()

ax.set_title('Anual revenue growth dynamics from 2010 to 2015', fontsize=20)
ax.set_ylabel('Revenue GR index, %', fontsize=16)

ax.set_yticklabels([f"{x:.0%}" for x in vals])

ax.legend(bbox_to_anchor=(1.02, 0.6), fontsize=16)

plt.show()
```

```python
# add graph with anual revenue dynamics by years

fig, ax = plt.subplots(figsize=(12, 6))

x = revenue.columns
companies = revenue.index

for num, elem in enumerate(companies):
    y = revenue.loc[elem, :]
    if elem == 'Shmuber':
        ax.plot(x, y, label=elem, linewidth=4)
    else:
        ax.plot(x, y, label=elem)
        
ax.set_title('Anual revenue dynamics from 2008 to 2015 by companies', fontsize=20)
ax.set_ylabel('Revenue, $K', fontsize=16)

ax.legend(bbox_to_anchor=(1.02, 0.6), fontsize=16)

plt.show()
```

## work with data for suitable visualisation

```python
# load excel and create df

market_lead = pd.read_excel('Домашнее_задание_2.xlsx', sheet_name='задание 2')
```

```python
market_lead.columns = market_lead.iloc[0, :]                 # rename df columns by first row data
market_lead = market_lead.drop(index=[0, 8, 9], axis=0)      # drop not unnecessary data
```

```python
market_lead.sort_values('share, %', inplace=True) # sort values by descending for correst barh visualisation
```

```python
market_lead
```

```python
# create colors dictionary and add mark companies by colors

green_hue = plt.get_cmap('Greens')(np.linspace(0.2, 1, len(market_lead['company'])))

colors_dict = dict(zip(pd.unique(market_lead['company']), green_hue))

market_lead['clrs'] = market_lead['company'].apply(lambda x: colors_dict[x])
```

```python
# create new dataset with total revenue by leaders and others 
lead_sum = market_lead[market_lead['company'] != 'others'].iloc[:, 1].sum()     # calculate lead total revenue
others_sum = market_lead[market_lead['company'] == 'others'].iloc[:, 1].sum()   # calculate others total revenue

pie_df = pd.DataFrame(dict(zip(['company_type', 'total_revenue'], \
                               [['lead', 'others'], [lead_sum, others_sum]])))  # create new df
```

```python
# add collors to companies type

colors_dict = dict(zip(pie_df['company_type'], (green_hue[6], 'white')))

pie_df['clrs'] = pie_df['company_type'].apply(lambda x: colors_dict[x])
```

```python
# create complex graph for suitable visualisation

fig, axs = plt.subplots(figsize=(20, 8), ncols=2)

for row in range(1):
    for col in range(2):
        ax = axs[col]
        if col == 0:
            x = market_lead.loc[6:1, 'company']
            y = market_lead.loc[6:1, 'share, %']
            colors = market_lead.loc[6:1, 'clrs']
            companies = list(x)

            bar = ax.barh(x, y, color=colors, alpha=0.5)

            labels_leg = dict(zip(companies, bar))

            x_ticks = ax.get_xticks()

            ax.set_title('Companies revenue share, %', fontsize=20)

            ax.set_xticklabels([f"{i:.2%}" for i in x_ticks])
            ax.tick_params(axis='both', labelsize=16)

            ax.bar_label(bar, label_type='center', labels=[f"{i:.2%}" for i in y], fontsize=16)

            # ax.legend(labels_leg.values(), labels_leg.keys(), bbox_to_anchor=(1.4, 0.6), fontsize='x-large')
            
        else:
            x = pie_df.loc[:, 'total_revenue']
            comp_type = list(pie_df.loc[:, 'company_type'])
            colors = pie_df.loc[:, 'clrs']

            patch, _, _ = ax.pie(x, labels=comp_type, 
                                 colors=colors, 
                                 autopct='%1.2f%%', 
                                 wedgeprops={"edgecolor": "black"}, 
                                 textprops={'fontsize': 16})
            
            ax.set_title('Revenue share by leading companies ans others', fontsize=20)

            for elem in patch:
                elem.set(alpha=0.5)

plt.show()
```

## my conclusions:
* по диаграмме трудно определить, кто лидер на рынке
* решил отобразить данные в виде barh для удобной визуализации лидеров от компании с самой большой долей дохода до компании с самой маленькой
* долю других компаний решил отобразить в виде диаграммы, так как более удобно для понимания, например, какой потенциал рынка есть за счет других компаний, а также можно сделать вывод, что доля рынка лидеров почти 40%
